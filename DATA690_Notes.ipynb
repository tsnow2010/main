```python
df["8. My Birth Year"]
```

```python
years = df["8. My Birth Year"].tolist()

acc_total = 0
acc_num = 0

for year in years:
  fixed_year = year[-4:]
  acc_total += int(fixed_year)
  acc_num += 1

print(int(2024 - acc_total/acc_num))

# Data was cleaned in that column!  (Intro to cleaning data!)
```
